## Phenomenon Being Modeled

Our project models the destruction of civilian and critical infrastructure in Ukraine as a result of Russia’s full-scale invasion that began in February 2022. Specifically, we are studying how the intensity and location of violent conflict events relate to the amount and type of infrastructure damage reported across Ukrainian regions. This phenomenon is significant because the scale of damage has directly shaped Ukraine’s humanitarian crisis, economic stability, and recovery priorities.

## Backround

Since 2022, the war has devastated Ukraine’s built environment. According to the World Bank’s 2024 Rapid Damage and Needs Assessment, total reconstruction costs were estimated at roughly $486 billion, with nearly half related to repairing or replacing infrastructure. The Ukrainian Ministry for Communities, Territories and Infrastructure Development reported that over 170,000 residential buildings have been damaged or destroyed, displacing millions of citizens. These losses have affected not only homes but also energy systems, hospitals, schools, and transportation networks, essential services that sustain everyday life.

For this analysis, we are using two main datasets. The Armed Conflict Location & Event Data Project (ACLED) Ukraine Conflict Monitor tracks weekly conflict events, noting where and when attacks occur and which types of infrastructure, energy, health, education, or residential were targeted. The Ukraine Damages dataset (GitHub, 2025) provides verified reports of what was physically damaged, the type of asset, and the status of repair. Together, these datasets allow us to link conflict activity to measurable outcomes of destruction.

## Main Features of Phenomenon 

Temporal Cycles of Damage

The data shows that infrastructure damage has followed distinct cycles tied to key phases of the war. For instance, damage levels spiked during the winters of 2022–2023 and 2023–2024 when Russian forces launched coordinated attacks on Ukraine’s energy grid. These periods correspond to large clusters of ACLED events tagged as “energy infrastructure,” reflecting a deliberate strategy to pressure civilians by disrupting power and heating.

Regional Concentration

Damage is not evenly distributed across the country. The eastern and southern oblasts—Donetsk, Luhansk, Zaporizhzhia, and Kherson—experience the highest density of destruction because they have been front-line areas for most of the conflict. Western regions, while still affected, show far fewer infrastructure-related events and damage reports.

Type of Infrastructure Affected

Residential housing makes up the majority of reported damages, followed by energy facilities and healthcare centers. According to ACLED data, about 60% of infrastructure-related attacks target homes or power systems. This pattern highlights how civilian spaces have become central to the conflict’s toll, both physically and socially.

Human and Economic Impact

The consequences extend beyond property loss. Damaged housing has displaced millions, destroyed schools have interrupted education for children, and the repeated targeting of hospitals and energy plants has put countless lives at risk. These realities underscore why modeling this phenomenon matters—it captures the measurable link between warfare and its effects on society.
